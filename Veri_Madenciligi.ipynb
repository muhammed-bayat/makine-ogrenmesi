## Karar Ağaçları


Ağaç tabanlı öğrenme algoritmaları, sıkça kullanılan gözetimli öğrenme algorimalarıdır. Genel olarak ele alınan bütün problemlerin (regresyon ve sınıflandırma ) çözümüne uyarlanabilir.
Her türlü veri bilimi problemlerinde karar ağaçları yaygın bir şekilde kullanılmaktadır.

**Karar ağaçları,sınıfı belirlenmiş ve geçmişte yaşanmış verilerden hareketle hangi sınıfa ait olduğu belirsiz olan verilerin ve sınıflarının belirlenmesini sağlayan bir tekniktir.**

Karar ağaçları, veri madenciliğinde tanımlayıcı ve tahminleyici niteliklere sahip olması,güvenilirlik düzeylerinin yüksek olması, değerlendirilmesinin ve uygulamasının kolay oluşu ve bilgisayar sistemlerindeki depolama ünitelerinde kolay entegre edilebilmeleri sebebiyle sıklıkla tercih edilmektedir. 



 Karar ağaçları, doğrusal olmayan ilişkilerde, aykırı değerlerin olduğu ve karma verilerin olduğu veri setlerinde sınıflandırma modelinin çıkarılmasına imkân verir.
 
Karar ağaçları, karar teorisi, sınıflandırma, tahmin ve kümelemede kullanılır.

 Hedef değişken sürekli ise karar ağaçları regresyon ağaçları olarak, kategorik ise sınıflama ağaçları olarak adlandırılmaktadır.Durum farklı olsa da, bir karar ağacı oluşturma yöntemi benzerdir.

![kararagac](https://miro.medium.com/max/500/1*McD6MUj47Gq71xAdvSTImg.png)

### Karar Ağaçlarının Avantajları
- Anlaması ve açıklaması kolaydır. Kullanılan ağaç yapı görselleştirilebilir. 
- Neredeyse hiç veri hazırlığı gerektirmez. Ancak, bu modelin kayıp değerleri desteklemediğine dikkat edilmelidir.
- Kullanılan ağacın maliyeti, ağacı eğitmek için kullanılan veri noktalarının sayısına göre logaritmiktir.
- Sayısal verileri ve sınıflandırılmış verileri işleyebilir.
- Birden fazla çıktılı problemleri çözebilirler.
- Modeli doğrulamak için istatistiksel testler kullanılabilir.
- Karar ağacı parametrik olmayan bir yöntem olarak düşünülebilir. Başka bir deyişle, sınıflandırma yapısı hakkında bir yaklaşıma  sahip değiller.

### Karar Ağaçlarının Dezavantajları
- Verileri iyi açıklamak için çok karmaşık bir ağaç oluşturabilir. Bu durumda ağaç dallanmas izlenemeyebilir.
- Ezbere öğrenme  (“over-fitting”) yaşanabilir. 
- Karar ağacı algoritmasında bölünmenin meydana gelme şekli, ağacın doğruluğunu etkileyen faktörlerden biridir.
- Entropi, verilerle ilişkili belirsizliğin bir ölçüsüdür. Bir veri kümesinin yalnızca bir etiketi varsa (örneğin, her yolcu hayatta kalırsa), entropisinin düşük olduğunu sezgisel olarak düşünebiliriz. Bu nedenle, verileri entropiyi en aza indirecek şekilde bölmemiz gerekiyor. Bölünme ne kadar iyi olursa tahminimiz o kadar iyi olur.




### Karar Ağaçları Nasıl Oluşturulur?

İstatistiksel bir yöntem olarak, karar ağacı yönteminin ilk adımı, bir ağaç yapısı oluşturmak ve veri kümesindeki verileri ağaca işlemektir. Karar ağacının yapısı kök düğümler, düğümler ve dallardan oluşur.

Kurallar oluşturulurken sorular sorulur ve karşı önlemler alınır. Cevaplar yeni bir kural oluşturmak için bir araya getirilir. Sorulacak ilk değişkene karar verdikten sonra, ağaç yapısının ilk adımı olan kök düğümü oluşturmaktır.

 Kök düğüm, bağımlı değişkeni temsil eder. Kök düğümden başlayarak ağaç yapısına göz atarken, veri seti küçük parçalara ve dallara bölünür ve yeni düğümler oluşturulur. Her düğüm bir soruyu temsil eder ve soruya verilen yanıt sayısı kadar dala bölünür. Her dalda bir "eğer-ise" kuralı oluşturulur ve bir sonraki düğüme seçilen karara göre geçilir. Ağaç yapısını yeni bir sorun ortaya çıkana kadar bu şekilde işlenir.En sonunda sınıfı temsil eden bir son düğüme ulaşılır.


## Karar Ağacı Parametreleri Nelerdir?
Karar ağacında hangi değişkenden dallanacağı önemli bir karardır. Çünkü mümkün olan tüm ağaç yapılarını oluşturmak ve işleme en uygun ağaç yapısından başlamak imkansızdır. Bu nedenle algoritma ilk aşamada bazı değerleri hesaplar ve sonuçlara göre ağaç oluşturma sürecini başlatır. Bu parametreler entropi, kazanç kriteri ve kazanç oranıdır.



## Karar Ağaçlarında Algoritma Seçimi

Algoritma, mevcut bir sorunu belirli bir sırayla veya belirli bir amaç için çözmek için mantıksal adımlar içeren bir çözümdür. Yani problemi çözmek için atılacak tüm adımlar algoritmayı oluşturmaktadır. Veri madenciliği araştırmalarında algoritmaların temel amacı, milyonlarca veri yığınında anlamlı modeller uygulamak için kurallar oluşturmaktır. Algoritma seçimi esas olarak modelin niteliklerine ve uygulanacak veri setine bağlıdır. Algoritma seçimi, modelin başarısını ciddi şekilde etkileyecektir. Bazı algoritmalar yalnızca nicel veriler üzerinde çalışırken, diğer algoritmalar yalnızca nitel veriler üzerinde çalışır. Diğerleri 0 ve l değerlerinden oluşan bir veri kümesi üzerinde çalışıyor. Yapay sinir ağı araştırmasında nitel değişkenlerin varlığı veya yokluğu olarak değerlendirilmesi veya karar ağacı araştırmasında nicel değişkenlerin kategorize edilmesi modelin performansını artırabilir.



Karar ağacı algoritmasının kullanıldığı sınıflandırma problemlerinde kullanılacak veri seti iki ana parçaya **(eğitim-train ve test)** ayrılmalıdır.

Algoritma, eğitim verilerini kullanarak model oluşturur. Oluşturulan bu model ile test verisi üzerinde uygulanarak modelin problem çözümündeki başarısı hesaplanır.

```python
import numpy as np 
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn import preprocessing 
from sklearn.model_selection import train_test_split

from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import accuracy_score

from sklearn.tree import DecisionTreeClassifier
```

>Veri setini tanımlıyoruz.

```python
data = pd.read_csv('games.csv')
```

>**train** ve **test** verilerini oluşturmak için x ve y tanımlıyoruz.

```python
x = data.drop(columns=['winner','id'])
y = data['winner']
```

>Text olarak geçen verileri sayısal değerlere dönüştürüyoruz.

```python
label_encoder = preprocessing.LabelEncoder()  
x = x.apply(label_encoder.fit_transform)
y= label_encoder.fit_transform(y)
```

>**train** ve **test** verilerini oluşturuyoruz.

```python
x_train, x_test, y_train, y_test = train_test_split(x, y)
```

>**Karar ağaçları** algoritmasını **train** verimiz ile çalıştırıp **model** oluşturuyoruz.

```python
clf = DecisionTreeClassifier()
clf = clf.fit(x_train,y_train)
```

>**test** verisi ile bir tahmin elde ediyoruz.

```python
y_pred = clf.predict(x_test)
```

>**Accuracy** skorumuzu hesaplıyoruz.

```python
print("Accuracy:",accuracy_score(y_test, y_pred))
```
